```python
# Name:- Majithiya Jal
# Enrollment NO.:- 216270307082
# Write a NumPy program to implement following operation
# to find the maximum and minimum value of a given flattened array
import numpy as np
flattened_array = np.arange(15)
np_array = np.array(flattened_array)
max_val = np.amax(np_array)
min_val = np.amin(np_array)
print(f"Maximum value: {max_val}")
print(f"Minimum value: {min_val}")
```
