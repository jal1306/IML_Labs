```python
# Name:- Majithiya Jal
# Enrollment NO.:- 216270307082
# Write a NumPy program to implement following operation
# to calculate mean across dimension, in a 2D numpy array
import numpy as np
x = np.array([[5,6], [7,8]])
print("Mean of x: ", np.mean(x))
print("Mean of a along axis=0: ", np.mean(x, axis=0))
print("Mean of a along axis=1: ", np.mean(x, axis=1))
```
