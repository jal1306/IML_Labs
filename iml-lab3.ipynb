```python
# Name:- Majithiya Jal
# Enrollment NO.:- 216270307082
# Write a NumPy program to implement following operation
# to stack arrays horizontally (column wise)

import numpy as np
a = np.array([[1,2,3],[4,5,6]])
b = np.array([[1, 2, 3],[4, 5, 6]])

print("\n a:- \n", a)
print("\n b:- \n", b)

c = np.hstack([a, b])
print("\nc: \n", c)
```
